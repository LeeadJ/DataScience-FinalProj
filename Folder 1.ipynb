The probability of P(Bowl 1|Chocolate) = $\frac{P(Bowl 1 \bigwedge Chocolate)}{P(Chocolate Cookies)}$ <br>


P(Bowl 1 $\bigwedge$ Chocolate)  =  P(Bowl 1) $\times$ P(Chocolate Cookies in Bowl 1)
<br>
P(Any given Bowl) = $\frac{1}{2}$ <br>
P(chocolate cookies in bowl 1) =  $\frac{30}{40} = \frac{3}{4}$ <br>

<br>
P(Chocolate Cookies) = P(Chocolate Cookies in Bowl 1) + P(Chocolate Cookies in Bowl 2)<br>
P(Chocolate Cookies in Bowl 1) = $\frac{3}{8}$<br>

###### Answer
